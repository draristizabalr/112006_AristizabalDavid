```python
#%% EJERCICIO 5
print('-----'*15)
print('EJEJRCICIO 5')
print('-----'*15)

import keyword as kw

texto = str(input('Escribir el texto que se va a analizar: '))

upper = 0
tilde = 0
digit = 0
reserv = 0
espace = 0

for i in texto:
    if int(str.isupper(i)) == 1:
        if i == 'A' or i == 'E' or i == 'I' or i == 'O' or i == 'U'\
 or i == 'Á' or i == 'É' or i == 'Í' or i == 'Ó' or i == 'Ú':
            upper = int(str.isupper(i)) + upper
    if i == 'á' or i == 'é' or i == 'í' or i == 'ó' or i == 'ú'\
 or i == 'Á' or i == 'É' or i == 'Í' or i == 'Ó' or i == 'Ú':
         tilde = tilde + 1
    digit = int(str.isdigit(i)) + digit
    espace = int(str.isspace(i)) + espace

palabras = str.split(texto)

for i in palabras:
    reserv = int(kw.iskeyword(i)) + reserv

    
print(f'La cantidad de vocales mayúsculas escritas en el texto es: {upper}')
print(f'La cantidad de vocales con tíldes escritas en el texto es: {tilde}')
print(f'La cantidad de digitos escritos en el texto es: {digit}')    
print(f'El numero de palabras reservadas que se escribieron fueron: {reserv}')
print(f'La cantidad de espacios en el escrito es: {espace}')
```
